```python
import pandas as pd
import numpy as np
from PIL import Image
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
df_word = pd.read_csv('word.csv')
df_word
```

```python
for ele in ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']:
    dict = {}
    word = df_word['word_'+ele].tolist()
    count = df_word[ele].tolist()
    for i in range(len(word)):
        dict[word[i]]=count[i]

    wc = WordCloud(background_color='white', max_words=200, contour_width=0, colormap='ocean',
                  width=800, height=400, min_font_size=10, font_path='Facile Sans.ttf').generate_from_frequencies(dict)
    wc.to_file('word_cloud_'+ele+'.png')
```
